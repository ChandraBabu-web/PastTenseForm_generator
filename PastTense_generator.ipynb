```python
# A sample program to build English Generator
# Author: D. Chandra Babu
# Date: 24/04/2022

import re
import sys#This is just to stop the execution 


      
#grouping the similar words to get patterns
irregular_verbs=['beat','beset','bet','broadcast','cast','cost','cut','forecast','hit','hurt','let','put','rain','read','rid','shed','shut','slit']
e_dropped=['hide','slide','bite']
t_added=['deal','hear','lean','leap','learn','mean']
i_a=['drink','forbid','give','ring','shrink','sing','sink','sit']
i_u=['cling','dig','fling','sling','slink']
i_o=['drive','ride','shine','shrive','smite','bestride']
d_t=['lend','rend','send','bend']
o_e=['blow','grow','hold','behold']
o_a=['come','become']
a_e=['fall','befall']
a_u=['hang']
e_o=['get','forget']
e_t=['lose']
u_a=['run']
k_d=['make']
e_e=['feed','bleed','breed','meet']
o_o=['choose','shoot']
ee_ee=['sleep','keep','creep']
eel=['feel','kneel']
lee=['flee']
eez=['freez']
a_dropped=['lead']
ie=['lie']
ied_added=['study']
ay_ew=['slay']
ee_aw=['see','foresee']
ell_old=['tell','sell','foretell']
ake_ook=['mistake','foresake','shake']
ell_elt=['smell','dwell']

pattern=r'e'
pattern_one=r'i'
pattern_two=r'd'
pattern_three=r'o'
pattern_four=r'a'
pattern_five=r'e'
pattern_six=r'u'
pattern_seven=r'k'
pattern_eight=r'ee'
pattern_nine=r'oo'
pattern_10=r'eep'
pattern_11=r'eel'
pattern_12=r'lee'
pattern_13=r'eez'
pattern_14=r'ie'
pattern_15=r'y'
pattern_16=r'ay'
pattern_17=r'ell'
pattern_18=r'ake'

no_pattern= {"grind":"ground","bind":"bound","bring":"brought","buy":"bought","lay":"laid","beget":"begot","go":"went","seek":"sought","leave":"left","interweave":"interwove","have":"had","eat":"ate","ear":"bore","catch":"caught","can":"could","break":"broke","bespeak":"bespoke","beseech":"besought","bereave":"bereft"}

conttt = "y"

while conttt == "y":
  word=input("enter the word :")
  for i in no_pattern.keys():
    if i==word:
      past_word=no_pattern.get(i)
      print("The past form of the",word,"is",past_word)
      sys.exit()
  if word in e_dropped:
    past_word=re.sub(pattern,'',word)
    print("the past form of the",word,"is",past_word) 
  elif word in ake_ook:
    past_word=re.sub(pattern_18,'ook',word)
    print("the past form of the",word,"is",past_word)
  elif word in ell_old:
    past_word=re.sub(pattern_17,'old',word)
    print("the past form of the",word,"is",past_word)
  elif word in ell_elt:
    past_word=re.sub(pattern_17,'elt',word)
    print("the past form of the",word,"is",past_word)
  elif word in ee_aw:
    past_word=re.sub(pattern_eight,'aw',word)
    print("the past form of the",word,"is",past_word)
  elif word in ay_ew:
    past_word=re.sub(pattern_16,'ew',word)
    print("the past form of the",word,"is",past_word)
  elif word in ied_added:
    past_word=re.sub(pattern_15,'i',word)
    past_word=past_word+'ed'
    print("the past tense form of the",word,"is",past_word)
  elif word in ie:
    past_word=re.sub(pattern_14,'ay',word)
    print("the past form of the",word,"is",past_word)
  elif word in a_dropped:
    past_word=re.sub(pattern_four,'',word)
    print("the past form of the",word,"is",past_word)
  elif word in eez:
    past_word=re.sub(pattern_13,'oze',word)
    print("the past form of the",word,"is",past_word)
  elif word in lee:
    past_word=re.sub(pattern_12,'led',word)
    print("the past form of the",word,"is",past_word)
  elif word in eel:
    past_word=re.sub(pattern_11,'elt',word)
    print("the past form of the",word,"is",past_word)
  elif word in ee_ee:
    past_word=re.sub(pattern_10,'ept',word)
    print("the past form of the",word,"is",past_word)
  #keep-kept
  #sleep-slept
  elif word in e_e:
    past_word=re.sub(pattern_eight,'e',word)
    print("the past form of the",word,"is",past_word)
  elif word in o_o:
    past_word=re.sub(pattern_nine,'o',word)
    print("the past form of the",word,"is",past_word)
  elif word in e_o:
    past_word=re.sub(pattern_five,'o',word)
    print("the past form of the",word,"is",past_word)
  elif word in e_t:
    past_word=re.sub(pattern_five,'t',word)
    print("the past form of the",word,"is",past_word) 
  elif word in i_a:
    past_word=re.sub(pattern_one,'a',word)
    print("the past form of the",word,"is",past_word) 
  elif word in o_e:
    past_word=re.sub(pattern_three,'e',word)
    print("the past form of the",word,"is",past_word)
  elif word in o_a:
    past_word=re.sub(pattern_three,'a',word)
    print("the past form of the",word,"is",past_word) 
  elif word in a_e:
    past_word=re.sub(pattern_four,'e',word)
    print("the past form of the",word,"is",past_word) 
  elif word in a_u:
    past_word=re.sub(pattern_four,'u',word)
    print("the past form of the",word,"is",past_word) 
  elif word in u_a:
    past_word=re.sub(pattern_six,'a',word)
    print("the past form of the",word,"is",past_word) 
  elif word in d_t:
    past_word=re.sub(pattern_two,'t',word)
    print("the past form of the",word,"is",past_word)  
  elif word in k_d:
    past_word=re.sub(pattern_seven,'d',word)
    print("the past form of the",word,"is",past_word)  
  elif word in i_u:
    past_word=re.sub(pattern_one,'u',word)
    print("the past form of the",word,"is",past_word) 
  elif word in i_o:
    past_word=re.sub(pattern_one,'o',word)
    print("the past form of the",word,"is",past_word) 
  elif re.search(r'([e])$', word):
    past_word=re.sub(r"([e])$", "\\1d",word)
    print("the past tense form of the",word,"is",past_word)
  #agree-agreed
  #admire-admired
  elif word in irregular_verbs:
      print("the past tense form of the",word,"is",word)
  #cut,put,hit,let,hurt,etc.
  elif word in t_added:
      past_word=word+'t'
      print("the past tense form of the",word,"is",past_word)
  elif re.search(r'([jt])([oe])([gp])$', word):
    past_word=re.sub(r"([jt])([oe])([gp])$","\\1\\2\\3\\3ed",word)
    print("the past form of the",word,"is",past_word)
  #step-stepped
  #stop-stopped
  #jog-jogged
  elif re.search(r'([ll|ee])$', word):
    past_word=re.sub(r"([ll|ee])$", "\\0t",word)
    print("the past tense form of the",word,"is",past_word)
  #dwell-dwelt
  #smell-smelt

  elif re.search(r'(ee\w+)', word):
    past_word=re.sub(r"(ee)","\\0e",word)
    print("the past form of the",word,"is",past_word)
  #bleed-bled
  #feed-fed
  #breed-bred
  #meet-met

  elif re.search(r'([pas])([a])([y])$', word):
    past_word=re.sub(r"([pas])([a])([y])$", "\\1aid",word)
    print("the past form of the",word,"is",past_word)
  #pay-paid
  #say-said

  else:
    past_word=word+"ed"
    print("the past form of the",word,"is",past_word)
  #work-worked
  #play-played
  conttt = input("Do you want to continue : y or n - ")
else:
  print("Thank you")

      
    
```
